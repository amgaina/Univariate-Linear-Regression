## This notebook contains Python code for a univariate linear regression task.  In addition to Python, we will also use Pandas, Numpy, and Matplotlib libraries. We will write Python code for the hypothesis, cost, and gradient descent functions. Next, we will train linear models with choices of w and b. Finally, we will evaluate our model using the test dataset. 

### For a univariate linear regression model, the hypothesis is a linear equation of the form (Slope-intercept form) below: 
#### ùë¶ÃÇ =‚Ñé(ùë•)=b+wx where b is the intercept, and w is the slope. Our hypothesis function will calculate the predicted values (or targets) for a given set of inputs (features) and w and b.

```python
import numpy as np;
import matplotlib.pyplot as plt;
import pandas as pd
%matplotlib inline
```

```python
# The hypotheses

def h(x, w, b):
    """
    calculates the predicted values for a given set of inputs.
    :param x: inputs (feature values) - Numpy array of floats 
    :param b: intercept - float
    :param w: slope - float
    :return: predicted targets - Numpy array of floats
    
    """
    return w*x + b
```

```python
# Cost Function
def cost(w, b, X, y):
    """
     calculates the total error for a given set of inputs and thetas.
    :param b: intercept - float
    :param w: slope - float
    :param X: inputs (feature values) - Numpy array of floats
    :param y: inputs (actual target values) - Numpy array of floats 
    :return: total error - floats
    
    """
    diff = (np.subtract((h(X, w, b)), np.asarray(y)))
    squared_diff = np.square(diff)
    return (squared_diff.sum()/(2 * len(X)))
    
    
    
```

```python
# Gradient Descent Function
def gradient(b, w, alpha, X, y):
    """
     calculates the gradient descent for a given set of inputs.
    :param w: intercept - float
    :param b: slope - float
    :param alpha: learning rate - float
    :param X: inputs (feature values) - Numpy array of floats
    :param y: inputs (actual target values) - Numpy array of floats 
    :return: calculated w and b - tuple of floats
    
    """
    my_array_0 = (np.subtract(h(X, w, b),np.asarray(y)))
    my_array_1 = np.matmul(np.transpose(my_array_0), X)
    
    b_temp = (alpha / len(X)) * my_array_0.sum()
    w_temp = (alpha / len(X)) * my_array_1.sum()
    
    new_b = b-b_temp
    new_w = w-w_temp
    
    return (new_b, new_w) 
```

```python
# Get the data
test_data = pd.read_csv("./test.csv")
X_test = pd.DataFrame(data = test_data['x'])
y = pd.DataFrame(data = test_data['y'])
X = X_test.to_numpy()
```

```python
# Initialize theta_0, theta_1 and the learning rate 
alpha = 0.0001
b = 0
w = 0

# Convergence threshold
my_diff = 0.000001

# Initial cost value
diff = cost(b, w, X, y)

# We will count the number of iterations
my_iter = 0

while diff >= my_diff:
    # Draw the current hypothesis and the training 
    # set to the x-y plane.
    # We will draw all the hypothesis lines so that
    # the reader can see how the hypothesis changes
    # with each iteration.
    my_line = np.linspace(0,100)
    z = b + w * (my_line)
    plt.scatter(X,y)
    plt.plot(my_line,z, 'r')
    
    # calculate initial cost value
    initial_cost = cost(b, w, X, y)
    
    # calculate and assign the new b and w values
    b, w = gradient(b, w, alpha, X, y)
    
    # calculate the consecutive cost value
    new_cost = cost(b, w, X, y)
    
    # calculate the difference between the consecutive
    # cost values
    diff = initial_cost - new_cost
    
    my_iter += 1
    
plt.xlabel("Square Foot of the house in ft2")
plt.ylabel("Cost of the house")
plt.title("Showing how the gradient descent function works as different input of the w and b.")
plt.show()    
```

```python
# Initialize b, w and the learning rate 
alpha = 0.0001
b = -10
w = -10

# Convergence threshold
my_diff = 0.000001

# Initial cost value
diff = cost(b, w, X, y)

# We will count the number of iterations
my_iter = 0

while diff >= my_diff:
    # calculate initial cost value
    initial_cost = cost(b, w, X, y)
    
    # calculate and assign the new b and w values
    b, w = gradient(b, w, alpha, X, y)
    
    # calculate the consecutive cost value
    new_cost = cost(b, w, X, y)
    
    # calculate the difference between the consecutive
    # cost values
    diff = initial_cost - new_cost
    
    # print for every 10000 iteration
    my_iter += 1

print()
print("Final Cost: {}".format(new_cost))
print("Diff: {}".format(diff))
print("Theta 0: {}".format(b))
print("Theta 1: {}".format(w))
print("Iteration: {}".format(my_iter))
print()

# Draw the final hypothesis and the training 
# set to the x-y plane.
# We will not draw all the hypothesis lines 
# because of the high iteration count.
my_line = np.linspace(0,100)
z = b + w * (my_line)
plt.scatter(X,y)
plt.plot(my_line,z, 'r')

plt.title("Univariate Linear Regression")
plt.xlabel("Square Foot of the house in ft2")
plt.ylabel("Cost of the house")
plt.show()    
```
